manipulasi value pada kolom 'product_category_name' dan'product_category_name_english'

```python
def preprocess_columns_string(df, col1, col2, btitle):
    """
    Preprocess the specified columns in the DataFrame.

    Parameters:
    - df (pd.DataFrame): DataFrame to be processed.
    - col1 (str): Name of the first column to be processed.
    - col2 (str): Name of the second column to be processed.

    Returns:
    - pd.DataFrame: Processed DataFrame.
    """
    # Ubah tipe data kolom 'col1' dan 'col2' menjadi string
    df[[col1, col2]] = df[[col1, col2]].astype(str)
    # Ubah format huruf pada kolom 'col1' dan 'col2' menjadi huruf kapital pada awal kata
    df[col1] = df[col1].str.title()
    if btitle:
        df[col2] = df[col2].str.title()

    # Ganti underscore (_) dengan spasi pada kolom 'col1' dan 'col2'
    df[col1] = df[col1].str.replace("_", " ")
    df[col2] = df[col2].str.replace("_", " ")

    return df

# Gunakan fungsi dengan parameter yang sesuai
df_product = preprocess_columns_string(df_product, 'product_category_name', 'product_category_name_english', True)

# Tampilkan lima baris pertama dari DataFrame yang telah diubah
df_product.head()
```

Membuat beberapa kolom baru yang menampilkan rentang hari antara beberapa event

```python
# Kolom rentang hari dari "order_purchase_timestamp" hingga "order_estimated_delivery_date"
df_orders_i['days_to_delivery_estimate'] = (df_orders_i['order_estimated_delivery_date'] - df_orders_i['order_purchase_timestamp']).dt.days

# Kolom rentang (menit) dari "order_purchase_timestamp" hingga "order_approved_at"
df_orders_i['minutes_to_approval'] = (df_orders_i['order_approved_at'] - df_orders_i['order_purchase_timestamp']).dt.total_seconds() / 60

# Kolom rentang hari dari "order_purchase_timestamp" hingga "order_delivered_carrier_date"
df_orders_i['days_to_carrier_delivery'] = (df_orders_i['order_delivered_carrier_date'] - df_orders_i['order_purchase_timestamp']).dt.days

# Kolom rentang hari dari "order_purchase_timestamp" hingga "order_delivered_customer_date"
df_orders_i['days_to_customer_delivery'] = (df_orders_i['order_delivered_customer_date'] - df_orders_i['order_purchase_timestamp']).dt.days
```

```python
# Grup dataset berdasarkan order_status
grouped_by_status = df_orders_i.groupby('order_status')

# Agregasi dengan menggunakan kolom-kolom baru
aggregated_data = grouped_by_status.agg({
    'days_to_delivery_estimate': 'mean',
    'minutes_to_approval': 'mean',
    'days_to_carrier_delivery': 'mean',
    'days_to_customer_delivery': 'mean'
}).reset_index()

# Tampilkan hasil agregasi
aggregated_data
```

Secara umum, pesanan yang telah disetujui (approved) memiliki rata-rata estimasi waktu pengiriman yang lebih lama dibandingkan dengan pesanan yang telah dikirim (shipped) dan telah diantar ke pelanggan (delivered). Status pesanan yang dibatalkan (canceled) menunjukkan rata-rata waktu persetujuan (approval) yang lebih singkat, mungkin karena proses pembatalan dapat dilakukan lebih cepat.

```python
def visual_penjualan(year, data):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

    # Plot Order Count
    axes[0].plot(data['order_purchase_timestamp'], data['order_count'], label='Order Count', marker='o')
    axes[0].set_xlabel('Month in {}'.format(year))
    axes[0].set_ylabel('Order Count')
    axes[0].set_title('Monthly Order Count in {}'.format(year))
    axes[0].tick_params(rotation=45)
    axes[0].legend()

    # Plot Total Transaction
    axes[1].plot(data['order_purchase_timestamp'], data['total_transaction'], label='Total Transaction', marker='o', color='orange')
    axes[1].set_xlabel('Month in {}'.format(year))
    axes[1].set_ylabel('Total Transaction')
    axes[1].set_title('Monthly Total Transaction in {}'.format(year))
    axes[1].tick_params(rotation=45)
    axes[1].legend()

    plt.tight_layout()
    plt.show()

# Plotting for each year
for year, data in yearly_groups:
    visual_penjualan(year, data)
```

```python
def add_region_column(df, state_column, region_column):
    southeast_area = ['SP', 'RJ', 'ES','MG']
    northeast_area = ['MA', 'PI', 'CE', 'RN', 'PE', 'PB', 'SE', 'AL', 'BA']
    north_area =  ['AM', 'RR', 'AP', 'PA', 'TO', 'RO', 'AC']
    cent_west_area = ['MT', 'GO', 'MS' ,'DF' ]
    south_area = ['SC', 'RS', 'PR']

    df.loc[df[state_column].isin(southeast_area), region_column] = 'Southeast'
    df.loc[df[state_column].isin(northeast_area), region_column] = 'Northeast'
    df.loc[df[state_column].isin(north_area), region_column] = 'North'
    df.loc[df[state_column].isin(cent_west_area), region_column] = 'Centwest'
    df.loc[df[state_column].isin(south_area), region_column] = 'South'

# Gunakan fungsi untuk menambahkan kolom 'customer_Region'
add_region_column(ecommerce_df_clean, 'customer_state', 'customer_Region')

# Gunakan fungsi untuk menambahkan kolom 'seller_Region'
add_region_column(ecommerce_df_clean, 'seller_state', 'seller_Region')
```

```python
customer_counts = ecommerce_df_clean.groupby('customer_Region')['customer_id'].nunique().sort_values(ascending=False)
seller_counts = ecommerce_df_clean.groupby('seller_Region')['seller_id'].nunique().sort_values(ascending=False)
print(customer_counts)
print("================")
print(seller_counts)

# Palet warna unik untuk setiap region
customer_palette = sns.color_palette("pastel", n_colors=len(customer_counts))
seller_palette = sns.color_palette("pastel", n_colors=len(seller_counts))

# Plotting dengan Seaborn
plt.figure(figsize=(14, 6))

# Plotting untuk pelanggan
plt.subplot(1, 2, 1)
sns.barplot(x=customer_counts.values, y=customer_counts.index, palette=customer_palette)
plt.title('Distribution of Customers by Region')
plt.xlabel('Number of Customers')

# Plotting untuk penjual
plt.subplot(1, 2, 2)
sns.barplot(x=seller_counts.values, y=seller_counts.index, palette=seller_palette)
plt.title('Distribution of Sellers by Region')
plt.xlabel('Number of Sellers')

plt.tight_layout()
plt.show()
```

### RFM analysis

RFM analysis merupakan salah satu metode yang umum digunakan untuk melakukan segmentasi pelanggan (mengelompokkan pelanggan ke dalam beberapa kategori) berdasarkan tiga parameter, yaitu recency, frequency, dan monetary.

- Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan transaksi.
- Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan transaksi.
- Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari pelanggan tersebut.

membuat sebuah DataFrame untuk menampung informasi terkait tiga parameter RFM

```python
rfm_df = ecommerce_df_clean.groupby(by="customer_id", as_index=False).agg({
    "order_purchase_timestamp": "max", # mengambil tanggal order terakhir
    "order_id": "count", # menghitung jumlah order
    "payment_value": "sum" # menghitung jumlah revenue yang dihasilkan
})
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]
rfm_df.head()
```

```python
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = df_orders_i["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)
rfm_df.head()
```

```python
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
# Ganti urutan kolom
rfm_df = rfm_df[['customer_id', 'recency', 'frequency', 'monetary']]
rfm_df.info()
```

```python
rfm_df.describe()
```

```python

# Adjust the figure size
plt.figure(figsize=(12, 10))

#Recency
plt.subplot(3, 1, 1)
# Create histplot
sns.histplot(rfm_df['recency'], kde=True, stat="density")
# Setting some visual options
plt.title('Recency distribution')
plt.xlabel(None)

#Frequency
plt.subplot(3, 1, 2)
# Create histplot
sns.histplot(rfm_df['frequency'], kde=True, stat="density")
# Setting some visual options
plt.title('Frequency distribution')
plt.xlabel(None)

#Monetary
plt.subplot(3, 1, 3)
# Create histplot
sns.histplot(rfm_df['monetary'], kde=True, stat="density")
# Setting some visual options
plt.title('Monetary distribution')
plt.xlabel(None)

plt.tight_layout()
```

membuat kolom baru yang berisi nilai skor RFM customer berdasarkan pengklasifikasian masing-masing parameter RFM
- Recency (R):
      pd.qcut(rfm_df['recency'], 4, ['4', '3', '2', '1']): Membagi kolom 'recency' menjadi 4 kuartil (quartiles) dengan menggunakan fungsi pd.qcut(). Quartiles ini diberi label '4', '3', '2', dan '1' untuk menunjukkan level keaktifan pelanggan, di mana '4' adalah yang paling baru (recency tinggi) dan '1' adalah yang paling lama (recency rendah).

- Frequency (F):
      pd.qcut(rfm_df['frequency'].rank(method='first'), 4, ['1', '2', '3', '4']): Menggunakan fungsi rank() untuk memberi peringkat (rank) pada nilai frequency. Kemudian, hasil peringkat tersebut dibagi menjadi 4 kuartil menggunakan pd.qcut(), dan diberi label '1', '2', '3', dan '4' untuk menunjukkan tingkat frekuensi, di mana '4' adalah yang paling sering.

- Monetary (M):
      pd.qcut(rfm_df['monetary'], 4, ['1', '2', '3', '4']): Mirip dengan Frequency, kolom 'monetary' juga dibagi menjadi 4 kuartil dengan pd.qcut() dan diberi label '1', '2', '3', dan '4' untuk menunjukkan tingkat nilai moneter, di mana '4' adalah yang memiliki nilai moneter tertinggi.

```python
rfm_df['R'] = pd.qcut(rfm_df['recency'], 4, ['4', '3', '2', '1'])
rfm_df['F'] = pd.qcut(rfm_df['frequency'].rank(method='first'), 4, ['1', '2', '3', '4'])
rfm_df['M'] = pd.qcut(rfm_df['monetary'], 4, ['1', '2', '3', '4'])
```

```python
# Menghitung RFM Score
rfm_df['RFM_Score'] = rfm_df['R'].astype('int') + rfm_df['F'].astype('int') + rfm_df['M'].astype('int')
# Menampilkan hasil
rfm_df.head()
```

    - RFM Score: 4-5
